```python
# 리팩토링 과제 힌트
# 구조는 아래와 같습니다.
class BungeoppangShop:
    def __init__(self):

    def check_stock(self):

    def process_order(self, bread_type, bread_count):

    def admin_mode(self):

    def calculate_total_sales(self):

def main():

if __name__ == "__main__":
    main()
```

과제_1: BungeoppangShop 클래스 만들기

- __init__: 초기 재고, 가격, 판매량을 설정합니다.
- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
- admin_mode: 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales: 총 매출을 계산하고 출력합니다.

```python
class BungeoppangShop:
    def __init__(self):
        # 초기 재고, 가격, 판매량 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        # 현재 재고 출력
        print("현재 재고:")
        for kind, count in self.stock.items():
            print(f"{kind}: {count}개")

    def process_order(self, bread_type, bread_count):
        # 주문 처리 및 재고 업데이트
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f"{bread_count}개의 {bread_type} 주문이 완료되었습니다.")
            else:
                print(f"{bread_type}의 재고가 부족합니다.")
        else:
            print(f"{bread_type}는 판매하지 않는 품목입니다.")

    def admin_mode(self):
        # 관리자 모드에서 재고 추가
        while True:
            kind = input("추가할 붕어빵의 종류를 입력하세요 (종료하려면 '종료' 입력): ")
            if kind == '종료':
                break
            
            if kind in self.stock:
                count = int(input(f"{kind} 몇 개 추가하시겠습니까?: "))
                self.stock[kind] += count
                print(f"{kind}의 재고가 {count}개 추가되었습니다.")
            else:
                print(f"{kind}는 판매하지 않는 품목입니다.")

            self.check_stock()  # 현재 재고 출력

    def calculate_total_sales(self):
        # 총 매출 계산 및 출력
        total_sales = sum(self.sales[kind] * self.prices[kind] for kind in self.sales)
        print(f"총 매출: {total_sales}원")


def main():
    shop = BungeoppangShop()

    # 예시: 주문 처리
    shop.process_order("팥붕어빵", 2)
    shop.process_order("슈크림붕어빵", 1)
    shop.check_stock()
    
    # 관리자 모드 예시
    shop.admin_mode()
    
    # 총 매출 출력
    shop.calculate_total_sales()

if __name__ == "__main__":
    main()
```

```python
class BungeoppangShop:
    def __init__(self):
        # 초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        
        # 가격 설정
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        
        # 판매량 초기화
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        # 현재 재고 출력
        print("현재 붕어빵 재고:")
        for kind, count in self.stock.items():
            print(f"{kind}: {count}개")

    def process_order(self, bread_type, bread_count):
        # 주문 처리 및 재고 업데이트
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f"{bread_count}개의 {bread_type} 주문이 완료되었습니다.")
            else:
                print(f"{bread_type}의 재고가 부족합니다. 현재 재고: {self.stock[bread_type]}개")
        else:
            print(f"{bread_type}는 판매하지 않는 품목입니다.")

    def admin_mode(self):
        # 관리자 모드에서 재고 추가
        while True:
            bread_type = input("추가할 붕어빵의 종류를 입력하세요 (종료하려면 '종료' 입력): ")
            if bread_type == '종료':
                break
            
            if bread_type in self.stock:
                try:
                    additional_stock = int(input(f"{bread_type} 몇 개 추가하시겠습니까?: "))
                    if additional_stock < 0:
                        print("추가할 재고는 0개 이상이어야 합니다.")
                    else:
                        self.stock[bread_type] += additional_stock
                        print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
                        self.check_stock()  # 현재 재고 출력
                except ValueError:
                    print("유효한 숫자를 입력하세요.")
            else:
                print(f"{bread_type}는 판매하지 않는 품목입니다.")

    def calculate_total_sales(self):
        # 총 매출 계산
        total_sales = 0
        for bread_type, quantity in self.sales.items():
            total_sales += quantity * self.prices[bread_type]
        
        print(f"총 매출: {total_sales}원")

# 클래스 테스트
if __name__ == "__main__":
    shop = BungeoppangShop()
    shop.check_stock()  # 현재 재고 확인
    shop.process_order("팥붕어빵", 2)  # 주문 처리
    shop.process_order("슈크림붕어빵", 1)  # 주문 처리
    shop.calculate_total_sales()  # 총 매출 계산 및 출력
```

- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
class BungeoppangShop:
    def __init__(self):
        # 초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        
        # 가격 설정
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        
        # 판매량 초기화
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        # 현재 재고 출력
        print("현재 붕어빵 재고:")
        for kind, count in self.stock.items():
            print(f"{kind}: {count}개")

    def process_order(self, bread_type, bread_count):
        # 주문 처리 및 재고 업데이트
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f"{bread_count}개의 {bread_type} 주문이 완료되었습니다.")
            else:
                print(f"{bread_type}의 재고가 부족합니다. 현재 재고: {self.stock[bread_type]}개")
        else:
            print(f"{bread_type}는 판매하지 않는 품목입니다.")

    def admin_mode(self):
        # 관리자 모드에서 재고 추가
        while True:
            bread_type = input("추가할 붕어빵의 종류를 입력하세요 (종료하려면 '종료' 입력): ")
            if bread_type == '종료':
                break
            
            if bread_type in self.stock:
                try:
                    additional_stock = int(input(f"{bread_type} 몇 개 추가하시겠습니까?: "))
                    if additional_stock < 0:
                        print("추가할 재고는 0개 이상이어야 합니다.")
                    else:
                        self.stock[bread_type] += additional_stock
                        print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
                        self.check_stock()  # 현재 재고 출력
                except ValueError:
                    print("유효한 숫자를 입력하세요.")
            else:
                print(f"{bread_type}는 판매하지 않는 품목입니다.")

    def calculate_total_sales(self):
        # 총 매출 계산
        total_sales = 0
        for bread_type, quantity in self.sales.items():
            total_sales += quantity * self.prices[bread_type]
        
        print(f"총 매출: {total_sales}원")

def main():
    shop = BungeoppangShop()

    while True:
        mode = input("모드를 선택하세요 ('주문' 또는 '관리자', 종료하려면 '종료' 입력): ")
        if mode == '종료':
            break
        elif mode == '주문':
            shop.check_stock()  # 현재 재고 확인
            bread_type = input("주문할 붕어빵의 종류를 입력하세요: ")
            try:
                bread_count = int(input("주문할 개수를 입력하세요: "))
                shop.process_order(bread_type, bread_count)  # 주문 처리
            except ValueError:
                print("유효한 숫자를 입력하세요.")
        elif mode == '관리자':
            shop.admin_mode()  # 관리자 모드
        else:
            print("잘못된 입력입니다. '주문' 또는 '관리자'를 입력하세요.")

    # 프로그램 종료 시 총 매출 계산 및 출력
    shop.calculate_total_sales()

# 프로그램 실행
if __name__ == "__main__":
    main()

```
